```python
import pandas as pd
import re
import matplotlib.pyplot as plt

# Load data
tweets = pd.read_csv('tweets.csv')

# Process tweet timestamps and extract Wordle ID & attempts
def process_tweets(tweets):
    tweets['tweet_datetime'] = pd.to_datetime(tweets['tweet_date'], errors='coerce')
    tweets['tweet_date'] = tweets['tweet_datetime'].dt.date

    pattern = r"Wordle (\d+) (\d)/6"
    extracted = tweets['tweet_text'].str.extract(pattern)

    tweets['wordle_id'] = extracted[0].astype('Int64')
    tweets['n_attempts'] = extracted[1].astype('Int64')

    return tweets

tweets = process_tweets(tweets)

print(tweets.head())
```

```python
import matplotlib.pyplot as plt

color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]

tweets['tweet_date'].value_counts().sort_index().plot(
    figsize=(10, 5),
    title="Wordle Tweets by Day",
    color=color_pal[2],
    lw=2
)
plt.xlabel("Date")
plt.ylabel("Number of Tweets")
plt.tight_layout()
plt.show()
```

```python
ax = tweets['n_attempts'].value_counts() \
    .sort_index() \
    .plot(figsize=(10, 5),
          kind='barh',
          title='Number of Wordle Attempts',
          edgecolor='black',
          color=color_pal[1])
ax.set_xlabel('Number of Tweets')
ax.set_ylabel('Number of Attempts Solved In')
plt.tight_layout()
plt.show()
```

```python
# Replace white squares with black squares for consistency
tweets['tweet_text'] = tweets['tweet_text'].str.replace('⬜', '⬛', regex=False)
```

```python
# Extract guesses and color information
def parse_tweet_text(tweets):
    for n in range(6):
        # Extract guess line
        tweets[f'has_guess{n+1}'] = tweets['tweet_text'].str.split('\n').str[n + 1] \
            .str.contains('|'.join(['🟨', '⬛', '🟩']), na=False)

        tweets.loc[tweets[f'has_guess{n+1}'], f'guess{n+1}'] = \
            tweets.loc[tweets[f'has_guess{n+1}'], 'tweet_text'].str.split('\n').str[n + 1].str[:5]

        tweets[f'guess{n+1}_black'] = tweets[f'guess{n+1}'].str.count('⬛')
        tweets[f'guess{n+1}_yellow'] = tweets[f'guess{n+1}'].str.count('🟨')
        tweets[f'guess{n+1}_green'] = tweets[f'guess{n+1}'].str.count('🟩')

        # Mark the final guess if all green
        tweets.loc[tweets[f'guess{n+1}_green'] == 5, 'final_guess'] = n + 1

    return tweets

tweets = parse_tweet_text(tweets)
```

```python
# Extract non-emoji last line (often comments)
tweets['additional_text'] = tweets['tweet_text'].str.split('\n').str[-1]
tweets['additional_text'] = tweets['additional_text'].where(~tweets['additional_text'].str.contains('🟩', na=False))
```

```python
# Define custom titles for each category
titles = ['Correct', 'Wrong Spot', 'Incorrect']
suffixes = ['_green', '_yellow', '_black']

fig, axs = plt.subplots(1, 3, figsize=(14, 5), sharex=True)

for i, suffix in enumerate(suffixes):
    col_subset = [c for c in tweets.columns if suffix in c]

    if not col_subset:
        print(f"No columns found for {suffix}")
        continue

    guess_avg = tweets[col_subset].mean()
    guess_avg.index = [f'Guess{j+1}' for j in range(len(guess_avg))]
    guess_avg = guess_avg.sort_index(ascending=False)

    guess_avg.plot(
        kind='barh',
        title=titles[i],  # Use clean title here
        ax=axs[i],
        color=color_pal[i + 1]
    )
    axs[i].set_xlabel('Average Letters')

fig.suptitle('Wordle Average Results by Guess Number', fontsize=18)
plt.tight_layout()
plt.show()
```
